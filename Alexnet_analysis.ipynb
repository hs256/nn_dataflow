```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
def return_total_cost(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['total_cost']
    return value
def return_total_time(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['total_time']
    return value
def return_total_op_cost(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['total_op_cost']
    return value
def return_total_noc_cost(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['total_noc_cost']
    return value
def return_total_access_cost(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['total_access_cost']
    return value

def return_active_node_pes(f):    
    with open(f, 'r') as file:
        data = json.load(file)
    value = data['active_node_pes']
    return value
```
